```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv(r"C:\Users\Akshay\Downloads\Superstoredata.csv",encoding="latin1")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop_duplicates()
df.head()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'],errors='coerce')
df['year_month']=df['Order Date'].dt.to_period('M')
df.head()
```

```python
df['Ship Date']=pd.to_datetime(df['Ship Date'].astype(str).str.replace('-','/'))
```

```python
df.head()
```

```python
df['Sales']=df['Sales'].round()
df['Profit']=df['Profit].astype(str
).str.replace('-','+')
df.head()
```

```python
df['Profit']=df['Profit'].round()
df.head()
```

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Region')[['Sales','Profit']].sum()
```

```python
df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
```

```python
df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

```python
df.groupby('year_month')['Sales'].sum()
```

```python
df.head(2)
```

```python
df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
#loss-analysis
df[df['Profit']<0].groupby('Sub-Category')['Profit'].sum()
```

```python
df['Ship Date'] = df['Ship Date'].fillna(df['Order Date'])
df.head(100)
```

```python
df = df.dropna(subset=['Order Date'])
df.head(100)
df.isnull().sum()
```

```python
df['delivery_days']=(df['Ship Date']-df['Order Date']).dt.days
df.head()
```

```python
df[['Sales','Discount','Profit']].corr()
df.groupby('Discount')['Profit'].mean()
```

```python
df.groupby('Discount')['Profit'].mean().sort_values()
#I found that higher discount percentages are negatively impacting profitability. Discounts above 30% consistently result in losses, suggesting a need to optimize discount strategies
```

```python
df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Profit')
```

```python
#correlation betweeen all values
df[['Sales','Profit','Discount','Quantity']].corr()
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```

```python
df.groupby('Sub-Category')['Profit'].sum().sort_values().plot(kind='barh')
plt.title("Profit by Sub-Category")
plt.show()
```

```python
df.groupby('year_month')['Sales'].sum().plot(kind='line')
plt.title('Monthly Sales Trend')
plt.show()
```

```python
plt.scatter(df['Discount'], df['Profit'])
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.title("Discount vs Profit")
plt.show()
```

```python
df.groupby('Ship Mode')['delivery_days'].mean().plot(kind='bar')
plt.title("Average Delivery Days by Ship Mode")
plt.show()
```

```python
df.to_csv("Superstore_data.csv",index=False)
```
